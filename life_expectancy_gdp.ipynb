```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# READING THE CSV

```python
df = pd.read_csv("all_data.csv")
```

# HEAD

```python
df.head()
```

# RENAMING COLUMN Life expectancy at birth (years) to Life expectancy

```python
df.rename(columns ={"Life expectancy at birth (years)": "Life expectancy"}, inplace = True)
```

# LETS GROUP BY Country AND Year TAKE THE MEAN Life expectancy

```python
df_Country_Year = df.groupby(["Country", "Year"])["Life expectancy"].mean()
```

# HEAD

```python
df_Country_Year.head()
```

# TURNING INTO PIVOT TABLE

```python
df_Country_Year_pivoted = pd.pivot_table(df, 
                                 index='Country', 
                                 columns='Year', 
                                 values='Life expectancy', 
                                 aggfunc='mean')
```

# HEAD


```python
df_Country_Year_pivoted.head()
```

# VISULATION 

```python
df_Country_Year_pivoted.T.plot(figsize=(12, 6))  
plt.title("Life Expectancy Over Time by Country")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.legend(loc='best')
plt.grid(True)
plt.show()
plt.clf()
```

# NOW LETS SEE THE  MEAN OF GPD OVER TIME

```python
df_Country_Year_GDP = df.groupby(["Country", "Year"])["GDP"].mean()
df_Country_Year_GDP.head()
```

# PIVOTED

```python
df_GDP_Year_pivoted = pd.pivot_table(df, index = "Country", columns = "Year", values = "GDP", aggfunc = "mean")
df_GDP_Year_pivoted.head()
```

# VISULATION OF THE GPD MEAN OVER THE TIME 

```python
df_GDP_Year_pivoted.T.plot(figsize = (12, 6))
plt.title("GDP Over Time Based on Country")
plt.xlabel("Year")
plt.ylabel("GDP")
plt.legend(loc = "best")
plt.grid(True)
plt.show()
```

# LETS SEEE IS THERE A RELATION SHIP WITH GDP AND LIFE Life Expectancy Over Time by Country

```python
sns.scatterplot(x = 'Life expectancy', y = 'GDP', hue='Country', data = df)
plt.show()
```
